# Prerequisite
1. `bin(x)`: Convert an `integer` number to a `binary string` prefixed with “0b”.
2. Bit Manipulation Operators only works on `int`

```
bin(2)
```

```
print(f"bin(10):\t {bin(10)}")
print(f"bin(-10):\t {bin(-10)}")
```

# `&`  And Operator

```
print(f"int('1010', 2):\t {int('1010', 2)}")
print(f"int('1001', 2):\t {int('1001', 2)}")

print("===================")
print(f"int('1000', 2):\t {int('1000', 2)}")
print(f"10 & 9:\t\t {10 & 9}")
```

## `|` Or Operator

```
print(f"int('1010', 2):\t {int('1010', 2)}")
print(f"int('1001', 2):\t {int('1001', 2)}")

print("===================")
print(f"int('1011', 2):\t {int('1011', 2)}")
print(f"10 | 9:\t\t {10 | 9}")
```

## `^` Exclusive-or, Xor Operator

```
print(f"int('1010', 2):\t {int('1010', 2)}")
print(f"int('1001', 2):\t {int('1001', 2)}")

print("===================")
print(f"int('0011', 2):\t {int('0011', 2)}")
print(f"10 ^ 9:\t\t {10 ^ 9}")
```

## `~` Not Operator

```
print(f"int('1010', 2):\t {int('1010', 2)}")
print(f"int('1001', 2):\t {int('1001', 2)}")

print("===================")
print(f"~ 10:\t\t {~10}")
print(f"0101:\t\t {int('0101', 2)}")
```

# Shift Operators
## `<<` left shift

```
a = 10
print(f"{bin(a)}:\t {a}")

a <<= 1  # a = a << 1
print(f"{bin(a)}: {a}")
```

```
a = 10
print(f"{bin(a)}:\t {a}")

a <<= 2
print(f"{bin(a)}: {a}")
```

## `>>` right shift

```
a = 10
print(f"{bin(a)}:\t {a}")

a >>= 2
print(f"{bin(a)}: {a}")
```
