```
def inifinite_sequence():
    num = 0
    while True:
        yield num
        num += 1
```

```
num_gen = inifinite_sequence()
num_gen
```

```
# Use for loop:
counter = 0
for i in num_gen:
    if counter == 5:
        break
    print(f"counter: {counter}, value: {i}")
    counter += 1
```

```
# call next(): python calls `.__next__()` on the functions
next(num_gen)
```

```
num_gen.__next__()
```

# [itertools.chain](https://docs.python.org/3/library/itertools.html#itertools.chain)

```
from itertools import chain
```

```
list(chain('ABC', 'DEF'))
```

```
def gen1():
    for i in range(3):
        yield i

def gen2():
    while True:
        yield -1

g1 = gen1()
g2 = gen2()
```

```
res = chain(*[g1, g2])
res
```

```
next(res)
```

```
next(res)
```

```
next(res)
```
